```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y,learning_rate,no_of_steps):
    m=0
    c=0
    nrows=len(x)
    for step in range(no_of_steps):
        y_cap=m*x+c
        cost_function=1/nrows* sum([val**2 for val in (y-y_cap)])
        dm=-2/nrows*sum(x*(y-y_cap))
        dc=-2/nrows*sum(y-y_cap)
        m=m-learning_rate*dm
        c=m-learning_rate*dc
        print(f'Step {step + 1} | Slope = {m} | Intercept = {c} | Cost = {cost_function}')
```

```python
x = np.array([1,2,3,4,5])
y = np.array([10,11,14,18,23])
gradient_descent(x,y,0.01,2000)
```
